```python
#zad 1
import math

bwzg1 = abs(math.pi - 22/7)
wzg1 = bwzg1/abs(math.pi)
print("Błąd bezwględny ", bwzg1, ", błąd względny: ", wzg1)

bwzg2 = abs(math.e - 2.718)
wzg2 = bwzg2 / abs(math.e)
print("Błąd bezwględny ", bwzg2, ", błąd względny: ", wzg2)

bwzg3 = abs(10**math.pi - 1397)
wzg3 = bwzg3 / abs(10**math.pi)
print("Błąd bezwględny ", bwzg3, ", błąd względny: ", wzg3)

bwzg4 = abs(math.factorial(9) - math.sqrt(18*math.pi)*(9/math.e)**9)
wzg4 = bwzg4 / abs(math.factorial(9))
print("Błąd bezwględny ", bwzg4, ", błąd względny: ", wzg4)

```

```python
#zad 2
import numpy as np
import matplotlib.pyplot as plt

pi = round(math.pi, 15)

G = lambda p : (1/p)*(10**p * (1 + p*pi*10**(-p)) - 10**p)
def ciag(n):
    list=[]
    for i in range (1,n+1):
        list.append(G(i))
    return list

print("math.pi: %.15f" % math.pi)
print("Wartość G(p) dla 20 elementów: ")
print(ciag(20))

#ODPOWIEDZI:
#1) nie zależy, p się redukuje
x = np.linspace(1,20)


plt.plot(x, [abs(G(i)-math.pi) for i in x])
plt.title('Wykres zależności błędu bezwzględnego od n')
plt.xlabel('n')
plt.ylabel('wartość błędu')
plt.yscale('log')
plt.show()

plt.plot(x, [abs((G(i)-math.pi)/math.pi) for i in x])
plt.title('Wykres zależności błędu względnego od n')
plt.xlabel('n')
plt.ylabel('wartość błędu')
plt.yscale('log')
plt.show()

#3) błąd jest na podobnym poziomie do n=16, dla n=17 błąd nieco się zwiększa, 
    #a od n=18 przybliżane wartości są całkowicie rozbieżne, gdyż różnią się od wartości dokładnej o 100%
#4) Występuje tu błąd nadmiaru i niedomiaru, dla liczb rzędu 10^18 i większych, 
    #wartości 10**(n) i 10**(-n) nie mogą być dokładnie obliczane, gdyż wykraczają poza zakres rejestru;
    #ponadto występuje błąd reprezentacji związany z zaokrąglaniem liczby niewymiernej pi
```
